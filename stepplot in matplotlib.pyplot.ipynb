```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(5)
y=np.arange(0,20,4)
plt.step(a,y,where="pre",linestyle="-",color="green",marker="*")
plt.title("step plot",color="red",style="italic",size=20)
plt.xlabel("X_values",color="blue")
plt.ylabel("Y_values",color="blue")
plt.show()
```
