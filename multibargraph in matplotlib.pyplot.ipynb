```python
import matplotlib.pyplot as plt
import numpy as np
classes=["6th","7th","8th","9th","10th"]
girls=[10,20,20,40,25]
boys=[20,30,25,30,20]
x_axis=np.arange(len(classes))
plt.bar(x_axis-0.2,girls,width=0.4)
plt.bar(x_axis+0.2,boys,width=0.4)
plt.xticks(x_axis,classes)
plt.xlabel("classes")
plt.ylabel("number of students")
plt.title("number of students in each class")
plt.legend(["male","female"])
plt.show()
```
