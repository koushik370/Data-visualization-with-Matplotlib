```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,0.1)
sin_values=np.sin(x)
cos_values=np.cos(x)
plt.plot(sin_values,color="blue",marker="^",markersize=5)
plt.plot(cos_values,color="k",marker="o",markersize=5)
plt.xlim(0,100)
plt.ylim(2,-2)
plt.title("sin and cos waves",color='red',fontfamily="cursive")
plt.xlabel("X_limits",color="blue",fontsize=14)
plt.ylabel("Y_limits",color="blue",fontsize=14)

plt.grid()
plt.show()
```
