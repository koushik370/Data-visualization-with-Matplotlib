```python
import matplotlib.pyplot as plt
name=['raju',"ramesh","sai","raghu","mohan","rakesh"]
salaris=[25000,27000,30000,35000,37000,40000]
plt.bar(name,salaris,width=0.6,edgecolor='black')
plt.title("employee monthly salaries",fontfamily="cursive",fontstyle='normal',fontsize=16,fontweight='bold',color='red')
plt.xlabel("employee names",color="blue",fontsize=14)
plt.ylabel("salaris range",color="blue",fontsize=14)
plt.grid()
plt.show()
```
