```python
import matplotlib.pyplot as plt
weight=(40,50,35,60,45,70,65)
plt.plot(weight,marker="*",markeredgecolor="red",markersize=5,mfc="green")
plt.title("markers in matplotlibs \n",color='red',fontsize=16)
plt.xlabel("num of students",fontdict={"family":"cursive","size":"25","color":"k"})
plt.ylabel("weight of students",fontdict={"family":"serif","size":"25","color":"k"})
plt.grid()
plt.show()
```
