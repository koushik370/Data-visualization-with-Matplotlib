```python
import matplotlib.pyplot as plt
departments=["cse","ece","civil","mechanical","eee"]
students=[28,35,19,25,10]
plt.bar(departments,students,color=["red","blue","green","cyan","purple"],edgecolor="k",label=departments,width=0.4)
plt.title("The number of students hired through placemnts",color="red",fontdict={"family":"serif","style":"italic","weight":"bold","size":14})
plt.xlabel("departments",color="olive",fontdict={"family":"monospace","weight":"bold","size":10})
plt.ylabel("number of students",color="olive",fontdict={"family":"monospace","weight":"bold","size":10})
plt.grid(axis="y",color="gray",linestyle="-.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show()


```
