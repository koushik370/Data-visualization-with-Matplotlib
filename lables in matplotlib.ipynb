```python
import matplotlib.pyplot as plt
x=("koushik","ramesh","rahul","karthik","rakesh","rithvik","remo")
y=(70,82,30,40,60,75,90)
plt.bar(x,y)
plt.xlabel("\n student names",fontdict={"family":"serif","size":"14","color":"blue"})
plt.ylabel("marks in english subject \n",fontdict={"family":"serif","size":"14","color":"blue"})
plt.title("marks sheet",color="red",fontsize=16)
plt.grid(axis="y")
plt.show()                                       
                                       
```
