```python
import matplotlib.pyplot as plt
t=[1,2,3,4]
a=[30,40,50,60]
b=[35,40,45,50]
c=[41,45,50,55]
labels_list=["a","b","c"]
colors_list=["blue","red","green"]
plt.stackplot(t,a,b,c,colors=colors_list,labels=labels_list)
plt.legend()
plt.title("Comparsion of marks")
plt.xlabel("months in year")
plt.ylabel("marks")
plt.show()
```
