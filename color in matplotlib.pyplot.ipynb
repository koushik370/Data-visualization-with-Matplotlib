```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(15)
y=x
plt.plot(x,y+0.4,color="green")
plt.plot(x,y+0.2,color="0.5")
plt.plot(x,y,color="red")
plt.plot(x,y-0.4,color="k")
plt.plot(x,y-0.2,color="m")
plt.plot(x,y-0.6,color="w")
plt.plot(x,y-0.8,color="blue")
plt.grid()
plt.title("color in matplotlib",color="red")
plt.show()



```
