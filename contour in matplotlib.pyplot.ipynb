```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-3,3,0.005)
y=np.arange(-3,3,0.005)
X,Y=np.meshgrid(x,y)
a=(X**2+Y**2)
out=plt.contour(X,Y,a)
plt.clabel(out,inline=True,fontsize=10)
plt.title("contour plot in matplotlib")
plt.xlabel("X_values")
plt.ylabel("Y_values")
plt.show()
```
