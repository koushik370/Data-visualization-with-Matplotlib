```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
a=["koushik","ramesh","sathish","bhasker"]
d=[20,60,35,40]
colors_names=["green",'k',"blue","yellow"]
plt.pie(d,labels=a,startangle=90,colors=colors_names,shadow=True)
plt.xlabel("name of the person")
plt.ylabel("monthly income of the person")
plt.legend(bbox_to_anchor=(1.5,1))
plt.title("pie chart example")
plt.show()
```
