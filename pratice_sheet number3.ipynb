```python
import matplotlib.pyplot as plt
import numpy as np
c=["7th","8th","9th","10th"]
boys=[27,30,35,42]
girls=[15,17,25,30]
x=np.arange(len(c))
plt.bar(x-0.2,boys,color="k",width=0.4,edgecolor="blue",label=boys)
plt.bar(x+0.2,girls,color="pink",width=0.4,edgecolor="blue",label=boys)
plt.xlabel("Number of students in diffent classes",size=16,family="serif")
plt.ylabel("Number of students in respective classes",size=16,family="serif")
plt.xticks(x,c)
plt.title("Total number of students in every class",size=20,family="cursive")
plt.grid(axis="y",linestyle="-.")
plt.show()
```
