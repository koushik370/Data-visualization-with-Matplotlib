```python
import matplotlib.pyplot as plt
days=["monday","tuesday","wednesday","thursday","friday","saturaday","sunday"]
temperature=[20,21,22,23,24,25,26]
humidity=[10,11,12,13,14,15,16]
plt.plot(days,temperature,marker="o",mfc="red",markeredgecolor="red")
plt.plot(days,humidity,marker="o",mfc="blue",markeredgecolor="blue")
plt.xlabel("Days in a week",size="14",color="k",family="cursive")
plt.ylabel("Temperature and humidity deatils",size=14,color="k",family="cursive")
plt.legend(["temperature","humidity"],bbox_to_anchor=(1,1))
plt.title("Details about temperature and humidity in this week",color="yellow",family="serif")
plt.xticks(ticks=days,rotation=45,color='red')
plt.grid()
plt.show()
```
