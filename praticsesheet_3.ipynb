```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.0,2.0,0.1)
y=1+np.sin(2*np.pi*x)
plt.fill(x,y,"k")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("fill plot example")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["car","bus","bikes","cycles"]
y=[50,30,60,15]
plt.step(x,y,where="mid",marker="o",linestyle="-.")
plt.xlabel("List of items in stock market")
plt.ylabel("percentage of items in stockmarket")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(size=(0,500))
y=np.random.normal(size=(0,500))
plt.hexbin(x,y,gridsize=20,mincnt=1, edgecolor="red",cmap="inferno")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```
