```python
import matplotlib.pyplot as plt
car=["audi","bmw","tesla","mercedes","tata"]
data=[23,17,35,29,12]
fig=plt.figure(figsize=(5,4))
plt.pie(data,labels=car,startangle=90,explode=[0,0.1,0.2,0,0.1],shadow=True,colors=["green","r","pink","b","k"],autopct="%1.2f%%")
plt.legend(title="car names",fontsize=8,bbox_to_anchor=(1.5,1))
plt.title("cars sale in 2022",color="r",fontsize=18)
plt.show()
        
```
