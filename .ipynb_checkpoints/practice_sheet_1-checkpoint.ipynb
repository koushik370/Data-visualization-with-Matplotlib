```python
import matplotlib.pyplot as plt
student_names=["koushik","anji","prithvi","karthik"]
marks=[68,55,60,36]
plt.bar(student_names,marks,color="red")
plt.xlabel("Name of students",fontdict={"family":"cursive","color":"blue","size":"14"})
plt.ylabel("Marks of student",fontdict={"family":"cursive","color":"blue","size":"14"})
plt.title("Semmister end examination marks",loc="center",size="14",color="red")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(6)
plt.plot(x,marker="*",markersize=12,markeredgecolor="red",mfc="green")
plt.xlabel("example of markeres",fontdict={"family":"cursive","color":"red",'size':"14"})
plt.title("nothing",loc="right",size='20',color="k")
plt.show()
```
