```python
import matplotlib.pyplot as plt
apples=[10,20,30,40,50,60]
bananas=[12,23,34,45,56,67]
plt.plot(apples,label="apples",marker="*",markersize=10)
plt.plot(bananas,label="bananas",marker="o",markersize=10)
plt.xlabel("number of apples sold",fontdict={"family":'serif',"color":"k","size":12})
plt.ylabel("number of bananas sold",fontdict={'family':"serif","color":"k",'size':12})
plt.title("Graph of selling of the items in peek stage",size=14,color="red")
plt.legend(fontsize=14,labelcolor='red',edgecolor="green",facecolor="w",shadow=True,title="legend box",bbox_to_anchor=(1,1))
plt.grid(axis="y")
plt.show()
```
