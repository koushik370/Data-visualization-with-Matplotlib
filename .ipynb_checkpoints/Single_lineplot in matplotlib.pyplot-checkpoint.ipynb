```python
import matplotlib.pyplot as plt
days=['monday','tuesday','wednessday','thursday','friday','staurday','sunday']
temperature=[20,21,22.5,23,23.5,25,25.5]
plt.plot(days,temperature,marker="*",markersize=10,mfc="red",markeredgecolor="red")
plt.legend(["days","temperature"],bbox_to_anchor=(1,1))
plt.xlabel("weeks",color='k',size='14')
plt.ylabel("Temperature analysis in a week", color="k",size="14")
plt.xticks(ticks=days,rotation=45,color="red")
plt.title("Wethear report of the week",color="green",size="16")
plt.grid()
plt.show()
```
