```python
import matplotlib.pyplot as plt
months=["january","feburay","march","april","may","june","july","augest","september","october","november","december"]
attendance=[81,85,75,73,78,79,85,83,84,83,75.5,76]
days=[90,100,80,85,92,80,84,90,94,77,90,99]
plt.scatter(months,attendance,color="b")
plt.scatter(months,days,color='k')
plt.title("Displaying the data of your attendence",color="red",size=16)
plt.xlabel("Months in year",color="green",size=14)
plt.ylabel("Attandance and total days",size=14,color="b")
plt.legend(["attendence","days"],bbox_to_anchor=(1,1))
plt.ylim(0,130)
plt.grid()
plt.show()



```
