```python
import matplotlib.pyplot as plt
party_names=["TRS",'BJP',"CONGRESS",'BSP','BRS']
votes=[132,124,104,118,100]
plt.bar(party_names,votes,width=0.4,edgecolor='black')
plt.title("analyzing diffent political parties \n",color="red",fontsize=15,fontweight="bold",fontfamily="cursive")
plt.grid(axis="y")
plt.xlabel("party names",color="blue",fontsize=14)
plt.ylabel("number of votes",color='blue',fontsize=14)
plt.show()
```
