```python
import matplotlib.pyplot as plt
import numpy as np
mobile_sales=np.random.normal(100,17,250)
Transcations=np.random.normal(140,20,25)
profit=np.random.normal(150,25,250)
loss=np.random.normal(160,15,250)
data=[mobile_sales,Transcations,profit,loss]
color=["pink","yellow","red","k"]
plt.violinplot(data,showmedians=True)
plt.xticks([1,2,3,4],['mobile_sales',"Transcations",'profit',"loss"])
plt.title("violin plot in the matplotlib",color="red")
plt.xlabel("variables",color="green")
plt.ylabel("data_values",color="blue")
plt.legend(["mobile_sales","Transcations","profit","loss"])
plt.show()
```
