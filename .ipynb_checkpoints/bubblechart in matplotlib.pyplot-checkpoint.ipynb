```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(100)
y=np.random.rand(100)
bubble_size=np.random.rand(100)
colors=np.random.rand(100)
plt.scatter(x,y,s=bubble_size*1000,c=colors)
plt.xlabel("X_values")
plt.ylabel("y_values")
plt.grid()
plt.title("Bubble chart in matplotlib",size=18,color="red")
plt.show()
```
