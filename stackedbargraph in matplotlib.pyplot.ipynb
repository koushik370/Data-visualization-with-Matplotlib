```python
import matplotlib.pyplot as plt
import numpy as np
teams=["team a", "team b","tema c",'team d']
y1=np.array([10,20,10,30])
y2=np.array([20,25,15,25])
y3=np.array([12,15,19,16])
y4=np.array([10,29,13,19])
plt.bar(teams,y1,color="r")
plt.bar(teams,y2,bottom=y1,color="b")
plt.bar(teams,y3,bottom=y1+y2,color="y")
plt.bar(teams,y4,bottom=y1+y2+y3,color='g')
plt.xlabel("Teams")
plt.ylabel("Scores")
plt.legend(["round1","round2","round3","round4"])
plt.title("Score by teams in 4 rounds")
plt.show()
```
