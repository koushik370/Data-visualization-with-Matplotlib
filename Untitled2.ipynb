```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame(np.random.rand(10,5),columns=["a","b","c","d","e"])
color=dict(boxes="red",whiskers="yellow",caps="Darkorange",medians="k")
df.plot.box(color=color)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("box plot")
plt.grid()
plt.show()
```
