```python
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
b=[30,40,20,40,50]
c=[35,20,50,30,80]
d=[90,70,80,40,70]
plt.subplot(2,2,1)
plt.plot(a,b)
plt.title("line plot",color="blue")
plt.grid()
plt.title("line plot",color="blue")
plt.grid()
plt.subplot(2,2,2)
plt.stackplot(a,a,c)
plt.title("stack plot",color="blue")
plt.grid()
plt.subplot(2,2,3)
plt.scatter(a,c,color="blue")
plt.title("scatter plot",color="blue")
plt.grid()
plt.subplot(2,2,4)
plt.plot(a,d,b)
plt.title("multi line plot",color="blue")
plt.grid()
plt.legend(["d","b"],bbox_to_anchor=(1,1))
plt.suptitle("sub plot in matplotlib",color="red",fontweight="bold")
plt.show()
```
