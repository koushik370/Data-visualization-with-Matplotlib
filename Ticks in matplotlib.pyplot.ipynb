```python
import matplotlib.pyplot as plt
import numpy as np
names=["koushik","anji","prithvi","karthik","hari prakesh"]
heights=[5.6,5.5,5.7,6.0,6.2]
plt.plot(names,heights,marker="^",markersize=5,mfc="yellow")
plt.xticks(ticks=names,rotation=45,color="red")
plt.yticks(ticks=np.arange(0,7,0.5),color="blue")
plt.title("Heights of students",fontdict={"size":15,"family":"serif","color":"green"})
plt.xlabel("X-axis ticks",color="red",size=14)
plt.ylabel("Y-axis ticks",color="red",size=14)
plt.grid()
plt.show()
```
